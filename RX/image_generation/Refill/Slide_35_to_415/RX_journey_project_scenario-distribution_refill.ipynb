```python
def get_mean_timedelta_stats(df,timecolumn):
    try:
        df = df[df[timecolumn]!='NA']
        df = df[pd.notnull(df[timecolumn])]
        df = df[df[timecolumn]>=dt.timedelta(0)]  #add = 

        df['timevalue'] = pd.to_timedelta(df[timecolumn])

        timemean = statistics.mean(df['timevalue'].dt.total_seconds().tolist())
        tc = len(df)

        return pd.to_timedelta(round(timemean,2),'s')
    except:
        pass

def get_median_timedelta(df,timecolumn):
    try:
        df = df[df[timecolumn]!='NA']
        df = df[pd.notnull(df[timecolumn])]
        df = df[df[timecolumn]>=dt.timedelta(0)] #add = 

        meantimedelta =  statistics.median(df[timecolumn].values.tolist())

        return pd.to_timedelta(meantimedelta)
    except:
        pass
```

```python
def class_hours(x):
    if 0<=x<=4:
        return '0-4 hr'
    elif 4<x<=8:
        return '4-8 hr'
    elif 8<x<=12:
        return '8-12 hr'
    elif 12<x<=16:
        return '12-16 hr'
    elif 16<x<=20:
        return '16-20 hr'
    elif 20<x<=24:
        return '20-24 hr'
    elif x>24:
        return '>24 hr'
    else:
        return np.nan
    
def class_hours_1hr(x):
    if x<=1:
        return '0-1 hr'
    elif 1<x<=2:
        return '1-2 hr'
    elif 2<x<=3:
        return '8-12 hr'
    elif 3<x<=4:
        return '12-16 hr'
    elif 4<x<=5:
        return '16-20 hr'
    elif 5<x<=6:
        return '20-24 hr'
    else:
        return '>6 hr'
    
def class_days(x):
    if 0<=x<=24:
        return '1-day'
    elif 24<x<=48:
        return '2-days'
    elif 48<x<=72:
        return '3-days'
    elif 72<x<=96:
        return '4-days'
    elif 96<x<=120:
        return '5-days'
    elif 120<x<=144:
        return '6-days'
    elif 144<x<=168:
        return '7-days'
    elif x>168:
        return '>7 days'
    else:
        return np.nan

def class_days_3day(x):
    if x<=3:
        return '0-3-days'
    elif 3<x<=6:
        return '3-6-days'
    elif 6<x<=9:
        return '6-9-days'
    elif 9<x<=12:
        return '9-12-days'
    elif 12<x<=15:
        return '12-15-days'
    elif 15<x<=18:
        return '15-18-days'
    elif 18<x<=21:
        return '18-21-days'
    else:
        return '>21 days'
```
